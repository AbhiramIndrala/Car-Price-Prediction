```python
#importing all the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#1.Taking the dataset
car=pd.read_csv("car data.csv")
car
```

```python
#2.Exploratory Data Analysis
print(car.columns)
car.shape#Shape of the dataset
car.size#Size ofthe dataset
print("\n")
print(car.info())
print("\n")
print(car.describe())
print("\n")
print(car.groupby("Car_Name").size())
print(car.groupby("Selling_Price").mean())
```

```python
#Checking the number of missing values
print(car["Fuel_Type"].value_counts())
print(car["Seller_Type"].value_counts())
print(car["Transmission"].value_counts())
```

```python
#Encoding the fuel column
car.replace({"Fuel_Type":{"Petrol":0,"Diesel":1,"CNG":2}},inplace=True)
car.replace({"Seller_Type":{"Dealer":0,"Individual":1}},inplace=True)
car.replace({"Transmission":{"Manual":0,"Automatic":1}},inplace=True)
```

```python
#3.DataVisualization
car.drop(["Car_Name"],axis=1,inplace=True)
```

```python
sns.heatmap(car.corr(),cbar=True,cmap="Blues",linewidth=1)
```

```python
sns.pairplot(car,diag_kind="hist")
```

```python
plt.figure(figsize=(18,6))
plt.title("Selling price per year")
sns.histplot(x="Selling_Price",hue="Year",data=car,palette="RdYlGn")
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.title("Present Price per year")
sns.histplot(x="Present_Price",hue="Year",data=car,palette="Set1")
plt.show()
```

```python
x=car.drop(["Selling_Price"],axis=1)
y=car["Selling_Price"]
```

```python
#Train and test the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2)
```

```python
#LINEAR REGRESSION
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred1=model.predict(x_test)
```

```python
error_score=metrics.r2_score(y_test,pred1)
print("R2 Score",error_score*100)
```

```python
plt.scatter(y_test,pred1,color="red")
```

```python
#LASSO
Ls=Lasso()
```

```python
Ls.fit(x_train,y_train)
```

```python
pred2=Ls.predict(x_test)
```

```python
error_score=metrics.r2_score(y_test,pred2)
print("R2 Score",error_score*100)
```

```python
plt.scatter(y_test,pred2,color="green")
```

```python
DS=DecisionTreeRegressor()
DS.fit(x_train,y_train)
```

```python
pred3=DS.predict(x_test)
```

```python
error_score=metrics.r2_score(y_test,pred3)
print("R2 Score",error_score*100)
```

```python
plt.scatter(y_test,pred3,color="#808000")
```

```python
Rf=RandomForestRegressor()
```

```python
Rf.fit(x_train,y_train)
```

```python
pred4=Rf.predict(x_test)
```

```python
error_score=metrics.r2_score(y_test,pred4)
print("R2 Score",error_score*100)
```

```python
plt.scatter(y_test,pred4,color="purple")
```
